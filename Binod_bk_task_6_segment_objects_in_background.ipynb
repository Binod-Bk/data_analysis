```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Load the image
image = cv2.imread('input_a.png')  # Replace with your input file name
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)  # Convert to grayscale
```

```python
_, simple_thresh = cv2.threshold(gray, 127, 255, cv2.THRESH_BINARY)
```

```python
adaptive_thresh = cv2.adaptiveThreshold(
    gray, 255, cv2.ADAPTIVE_THRESH_MEAN_C, cv2.THRESH_BINARY, 11, 2
)
```

```python
_, otsu_thresh = cv2.threshold(gray, 0, 255, cv2.THRESH_BINARY + cv2.THRESH_OTSU)
```

```python
import mahotas
T = mahotas.thresholding.rc(gray)
riddler_thresh = gray.copy()
riddler_thresh[riddler_thresh > T] = 255
riddler_thresh[riddler_thresh <= T] = 0
```

```python
# Display the images
titles = ['Original Image', 'Simple Thresholding', 'Adaptive Thresholding', "Otsu's Thresholding", 'Riddler-Calvard']
images = [gray, simple_thresh, adaptive_thresh, otsu_thresh, riddler_thresh]

plt.figure(figsize=(12, 8))
for i in range(len(images)):
    plt.subplot(2, 3, i+1)
    plt.imshow(images[i], cmap='gray')
    plt.title(titles[i])
    plt.axis('off')
plt.tight_layout()
plt.show()
```

```python
cv2.imwrite('simple_thresh.jpg', simple_thresh)
cv2.imwrite('adaptive_thresh.jpg', adaptive_thresh)
cv2.imwrite('otsu_thresh.jpg', otsu_thresh)
cv2.imwrite('riddler_thresh.jpg', riddler_thresh)
```
